```python
import matplotlib.pyplot as plt
data=[1,2,3,4,5,6,7]
name=['s','d','e','r','t','a','l']
plt.bar(data,name,edgecolor="red")
plt.xlabel('data')
plt.ylabel('name')
plt.title('data')
plt.grid()
plt.show()

# Assuming you want to plot data against these names as x-axis ticks/labels
plt.figure(figsize=(8, 5))
plt.plot(name, data, marker='o')
plt.xlabel("Labels")
plt.ylabel("Values")
plt.title("Plot of Data")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
name=["ravi","naruto","minaro","itachi","sasuke","madara","boruto","hashirama"]
power=[99,98,85,97,90,92,86,89]
plt.scatter(name,power)
plt.xlabel("name")
plt.ylabel("power")
plt.title("name vs power")
plt.grid()
plt.show()

plt.pie(power, labels=name, autopct='%2.1f%%', startangle=90)
plt.title("Power Distribution")
plt.legend(name)
plt.show()
```

```python
import matplotlib.pyplot as plt
name=["ravi","naruto","minaro","itachi","sasuke","madara","boruto","hashirama"]
power=[99,98,85,97,90,92,86,89]
plt.hist(power, bins=5, edgecolor='black')
plt.xlabel("Power")
plt.ylabel("Frequency")
plt.title("Distribution of Power")
plt.show()
```

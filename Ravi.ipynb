```python
import pandas as pd
import numpy as np
data={
    "Product_ID": [100, 101, 102, 103, 104],
    "Product_Name": ["Laptop", "Mobile", "Mouse", "Headphone", "KeyBoard"],
    "Catergory": ["Electronics", "Monitor", "Accessories", "Accessorices", "Furniture"],
    "Price": [1000, 2000, 3000, 4000, 5000],
    "Quantity_Sold": [10, 20, 30, 40, 50]
}

#dataframe
df=pd.DataFrame(data)

#1.load and inspect of dataset
print("Dataset:")
print(df)

#2.handling missing
df.fillna(0,inplace=True)

#3.remove dupliactes
df.drop_duplicates(inplace=True)

#4.create revenue = price * quantity_sold
df["Revenue"]=df["Price"]*df["Quantity_Sold"]
print(df)

#5.products generating revenue above average
avg_revenue=df["Revenue"].mean()
print("\nProducts with Revenue Above Average:")
print(df[df["Revenue"]>avg_revenue])

#6.sort products by revenue
sorted_df=df.sort_values(by="Revenue",ascending=False)
print("\nSorted Dataset:")
print(sorted_df)

#7.category-wise total revenue
category_revenue=df.groupby("Catergory")["Revenue"].sum()
print("\nCategory wise Revenue:")
print(category_revenue)

#8.Numpy statistics on revenue
revenues=np.array(df["Revenue"])
print("\nRevenue Statistics:")
print("Mean Revenue:", np.mean(revenues))
print("Median Revenue:", np.median(revenues))
print("Maximum Revenue:", np.max(revenues))
print("Min Revenue:", np.min(revenues))
print("Standard Deviation:", np.std(revenues))

#9.Best_selling_product
best_selling_product=df[df["Quantity_Sold"]==df["Quantity_Sold"].max()]
print("\nBest Selling Product:")
print(best_selling_product)

#10Sales summary report
print("\nSALES SUMMARY REPORT:")
print("Total Revenue:", df["Revenue"].sum())
print("Total Quantity Sold:", df["Quantity_Sold"].sum())
print("Average Revenue Per Sale:", avg_revenue)
```

```python
import pandas as pd
import numpy as np

data = {
    "Product_ID": [100, 101, 102, 103, 104],
    "Product_Name": ["Laptop", "Mobile", "Mouse", "Headphone", "Keyboard"],
    "Category": ["Electronics", "Monitor", "Accessories", "Accessories", "Furniture"],
    "Price": [1000, 2000, 3000, 4000, 5000],
    "Quantity_Sold": [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data).fillna(0).drop_duplicates()

df["Revenue"] = df["Price"] * df["Quantity_Sold"]

print("Dataset:\n", df)

avg = df["Revenue"].mean()

print("\nAbove Average Revenue:\n", df[df["Revenue"] > avg])

print("\nSorted by Revenue:\n", df.sort_values("Revenue", ascending=False))

print("\nCategory Revenue:\n", df.groupby("Category")["Revenue"].sum())

print("\nStatistics:")
print("Mean:", np.mean(df["Revenue"]))
print("Median:", np.median(df["Revenue"]))
print("Max:", np.max(df["Revenue"]))
print("Min:", np.min(df["Revenue"]))
print("Std:", np.std(df["Revenue"]))

print("\nBest Selling Product:\n",
      df[df["Quantity_Sold"] == df["Quantity_Sold"].max()])

print("\nTotal Revenue:", df["Revenue"].sum())
print("Total Quantity Sold:", df["Quantity_Sold"].sum())
print("Average Revenue:", avg)
```
